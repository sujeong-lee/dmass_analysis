# DMASS-MG Results

### Unblinded by Sujeong Lee (Nov 12, 2020)

This work uses the 3x2pt data vectors using DMASS as lenses and Y1 metacal as sources.
Only the GGL measurement and relevant contours are blinded as Y1 metacal was already unblinded and the DMASS sample was defined to match its galaxy clustering with CMASS. The data vectors and contours are shifted randomly before plotting or numbers in the axes are hidden. 

In this notebook, we unblind the constraints on modified gravity constrained using the DMASS 3x2pt measurements and as data vectors. The unblinded GGL measurements can be find in LINK.

# Data vectors

Plot with the model cosmology illustrated in Intro section

```python
# measurement of 3x2pt 
fitsname = '../data/measurement_y1_dmass_3x2pt.fits'

# best fit theory 
theory_fits = '../analysis/results/bestfit_theory/bestfit_mg_y1_dmass_3x2pt_50bins.fits'
```

```python
import fitsio

fits = fitsio.FITS(fitsname)
data = fits['galaxy_xi']['VALUE'].read()
ang = fits['galaxy_xi']['ANG'].read()
bin_no = fits['galaxy_xi']['BIN2'].read()
cov = fits['COVMAT'].read()[-20:, -20:]
err = np.sqrt(cov.diagonal())

fits_theory = fitsio.FITS(theory_fits)
theory = fits_theory['galaxy_xi']['VALUE'].read()
ang_theory = fits_theory['galaxy_xi']['ANG'].read()
bin_no_theory = fits_theory['galaxy_xi']['BIN2'].read()
```

```python
xlabel = '$\\theta$ ${\\rm [arcmin]}$'
ylabel='$w (\\theta)$'

fig, ax = plt.subplots(figsize=(3.2,2.8) )
ax.plot(ang_theory, theory, ls='--', lw=1, color='black')
#ax.plot(ang_theory2, theory2, ls='--', color='black')
ax.errorbar(ang, data, yerr=err, fmt='o', markerfacecolor=None,  color='steelblue', capsize=5)
ax.set_xscale('log')
ax.set_yscale('log', fontsize=17)
ax.set_xlabel(xlabel,fontsize=17)
ax.set_xlim(2.3, 290)
ax.set_ylim(4e-04, 0.4)
#ax.set_ylim(5e-02, 2)
ax.tick_params(labelsize=15, which='both', direction='in')
ax.axvspan(1, 7, alpha=0.3, color='grey')
#ax.legend(loc=4)
ax.legend(loc=1, fontsize=15, frameon=False)
ax.set_ylabel(ylabel,fontsize=17)
plt.tight_layout()

figname = figdir+'wtheta.pdf'
fig.savefig(figname)
```

```python
figname = figdir+'gammat.pdf'
plotting_gammat_measurement_from_fits( fitsname=fitsname, theory=theory_fits, blind=Blind_switch, 
                            scalecut=True, thetagamma=False, 
                            yextent = None, data_labels = None, angcut=[17,35,44,35], 
                            ylabel='$\gamma_t (\\theta)$', figname=figname)
```
